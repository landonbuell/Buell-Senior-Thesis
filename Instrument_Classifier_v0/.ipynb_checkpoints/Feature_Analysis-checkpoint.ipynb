```python
FRAME = pd.read_csv('C:/Users/Landon/Documents/GitHub/Buell-Senior-Thesis/Instrument_Classifier_v0/extdata/Output.csv',
                  index_col=0)
FRAME.head(10)
```

```python
y = FRAME['Target']
y = y.to_numpy().astype(np.float32)
y
```

```python
def Plot_Features(DATA,X1,X2,Y,savename=''):
    """ Plot 2D feature space, color code by class """
    plt.figure(figsize=(16,8))
    plt.xlabel(str(X1),size=20,weight='bold')
    plt.ylabel(str(X2),size=20,weight='bold')
    plt.scatter(x=DATA[X1],y=DATA[X2],c=Y,cmap='jet',label=[0,1,2,3])
    plt.legend()
    plt.grid()
    plt.tight_layout()
    plt.savefig(savename+'.png')
    plt.show()
```

```python
# Plot RISE vs. DECAY
Plot_Features(FRAME,'Rise Time','Decay Time',Y=y,
             savename="Rise_Decay")
```
