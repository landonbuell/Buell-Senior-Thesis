```python
confMatA = np.array([[101,0,2,0],
                     [1, 81,1,2],
                     [4, 0,99,0],
                     [0, 3,0,112]])

confMatB = np.array([[41,10,20,9],
                     [10,21,51,63],
                     [40,21,13,90],
                     [32,38,8 ,18]])
```

```python
def PlotConfusion(X,namesave=""):
    plt.figure(figsize=(12,12))
    plt.ylabel("Actual Class",size=60,weight="bold")
    plt.xlabel("Predicted Class",size=60,weight="bold")
    plt.imshow(X,cmap=plt.cm.binary)
    
    d = 0.06
    for i in range(X.shape[0]):
        for j in range(X.shape[1]):
            plt.text(x=i,y=j,s=str(X[j,i]),size=60,ha='center',va='center',color='red')
    
    #plt.grid()
    plt.xticks(np.arange(X.shape[0]),fontsize=50)
    plt.yticks(np.arange(X.shape[0]),fontsize=50)
    plt.tight_layout()
    plt.savefig(namesave+".png")
    plt.show()
```

```python
PlotConfusion(confMatA,"StdConfA")
```

```python
PlotConfusion(confMatB,"StdConfB")
```
