```python
def PlotFoldsMetrics(X,namesave):
    """ Plot Metrics from Cross Validation """
    _labs = ["Accuracy","Precision","Recall","F1"]
    _xlab = ["Split " + str(x) for x in range(0,10)]
    
    _barWidth = 0.2
    _x1 = np.arange(0,10)
    _x2 = np.arange(0,10) + _barWidth
    _x3 = np.arange(0,10) + 2*_barWidth
    _x4 = np.arange(0,10) + 3*_barWidth
    
    plt.figure(figsize=(20,8))
    plt.ylabel("Score Value",size=40,weight='bold')
    
    plt.bar(x=_x1,height=X[:,0],width=_barWidth,label=_labs[0],edgecolor='white')
    plt.bar(x=_x2,height=X[:,1],width=_barWidth,label=_labs[1],edgecolor='white')
    plt.bar(x=_x3,height=X[:,2],width=_barWidth,label=_labs[2],edgecolor='white')
    plt.bar(x=_x4,height=X[:,3],width=_barWidth,label=_labs[3],edgecolor='white')

    plt.xticks(_x3-_barWidth/2,_xlab,size=25,weight='bold',rotation=10)
    plt.yticks(size=25,weight='bold')
    
    plt.legend(fontsize=25,loc='lower left')
    plt.grid()
    plt.tight_layout()
    plt.savefig(namesave+".png")
    plt.show()
```

```python
fileA = "BySplitXValAlphaCLF@2020.11.29.21.02.38.868620@ANALYSIS.csv"

dataA = pd.read_csv(fileA,index_col=0)
dataA = dataA.to_numpy()
print(dataA.shape)

PlotFoldsMetrics(dataA,"AvgXval-Multimodal")
```

```python
fileB = "BySplitXValBetaMLP@2020.12.05.23.03.48.087052@ANALYSIS.csv"

dataB = pd.read_csv(fileB,index_col=0)
dataB = dataB.to_numpy()
print(dataB.shape)

PlotFoldsMetrics(dataB,"AvgXval-Perceptron")
```

```python
fileC = "BySplitXValGammaCNN@2020.12.05.23.04.33.366495@ANALYSIS.csv"

dataC = pd.read_csv(fileC,index_col=0)
dataC = dataC.to_numpy()
print(dataC.shape)

PlotFoldsMetrics(dataC,"AvgXval-Convolution")
```
