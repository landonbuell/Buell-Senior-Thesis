```python
filename = "SynthClassifier@2020.11.29.21.45.41.132119@TRAINING-HISTORY.csv"
filedata = pd.read_csv(filename)
filedata
```

```python
loss = filedata["Loss Score"]
prec = filedata["Precision"]
recl = filedata["Recall"]
epch = np.arange(len(loss))
```

```python
def PlotLossHistory(X,Y,namesave):
    """ Plot Loss at Each Epoch """
    plt.figure(figsize=(20,8))
    plt.xlabel("Epoch Number",size=40,weight='bold')
    plt.ylabel("Loss Score",size=40,weight='bold')
    
    plt.plot(X,Y,color='blue',linewidth=4,label="Loss History")
    
    plt.hlines(0,0,X[-1],color='black')
    plt.vlines(0,0,4,color='black')
    
    plt.xticks(size=20,weight='bold')
    plt.yticks(size=20,weight='bold')
    
    plt.grid()
    plt.legend(fontsize=25)
    plt.tight_layout()
    plt.savefig(namesave+".png")
    plt.show()
    plt.close()
    
PlotLossHistory(epch,loss,"LossHistory")
```

```python
def PlotMetricHistory(X,Y,Z,namesave):
    """ Plot Loss at Each Epoch """
    plt.figure(figsize=(20,8))
    plt.xlabel("Epoch Number",size=40,weight='bold')
    plt.ylabel("Metric Score",size=40,weight='bold')
    
    plt.plot(X,Y,color='purple',linewidth=4,label="Precision")
    plt.plot(X,Z,color='orange',linewidth=4,label="Recall")
    
    plt.hlines(0,0,X[-1],color='black')
    plt.vlines(0,0,1,color='black')
    
    plt.xticks(size=20,weight='bold')
    plt.yticks(size=20,weight='bold')
    
    plt.grid()
    plt.legend(fontsize=25,loc='lower right')
    plt.tight_layout()
    plt.savefig(namesave+".png")
    plt.show()
    plt.close()
    
PlotMetricHistory(epch,prec,recl,"MetricHistory")
```
