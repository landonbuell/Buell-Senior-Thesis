```python
yHat = np.arange(0.01,1,0.01)

CXE = -np.log10(yHat)

plt.figure(figsize=(20,8))
plt.ylabel("Cross-Entropy Loss Value",weight='bold',size=30)
plt.xlabel("Prediction Value",weight='bold',size=30)
plt.plot(yHat,CXE,linewidth=4,color='purple')
plt.hlines(0,0,1,color='black')
plt.vlines(0,0,2.5,color='black')
plt.xticks(size=20)
plt.yticks(size=20)
plt.grid()
plt.tight_layout()
plt.savefig("..\\Figures\\CXELoss")
plt.show()
```
