```python
import numpy as np
import matplotlib.pyplot as plt
import os

import scipy.fftpack as fftpack
import scipy.signal as signal

intdir = 'C:/Users/Landon/Documents/Github/Buell-Senior-Thesis/read_wavs'
bindir = 'C:/Users/Landon/Documents/wav_data/time_series'
```

```python
def read_directory(dir):
    """
    Read all files in given directory path
    --------------------------------
    dir (str) : Parent directory path to read raw .wav files from
    --------------------------------
    return list of "wav_file" class object instances
    """
    filenames = []                          # list to '.wav' hold files objs
    for roots,dirs,files in os.walk(dir):   # all objects in parent path
        for file in files:                  # files in list of files
            if file.endswith('.bin'):       # if '.wav' file
                filenames.append(file)      # add to list
    return filenames                        # return the list of files

files = read_directory(bindir)
```

```python
os.chdir(bindir)
M = 2**12

file = files[120]
print(file)

X = np.fromfile(file)

X = X.reshape(-1,M)

print(np.shape(X))
    
```

```python
wave = np.array([X[10],X[11]]).flatten()
plt.plot(wave)
plt.show()
```

```python
snip = np.array(X[16]).flatten()
fftdata = fftpack.fft(snip,n=M)
pwr = np.real(fftdata)**2

f = fftpack.fftfreq(n=M,d=(1/44100))

pts = np.where((f>0)&(f<4000))
fspace = f[pts]
amp = pwr[pts]
amp = amp/np.abs(np.max(amp))
```

```python
plt.figure(figsize=(20,8))
plt.plot(fspace,amp)
plt.show()

maxamp = np.argmax(amp)
print(fspace[maxamp])
```

```python
peaks = signal.find_peaks(amp,height=0.001,)
print(len(peaks[0]))

print(peaks[0])
```

```python
plt.figure(figsize=(20,8))
plt.plot(fspace,amp,color='blue')
plt.scatter(fspace[peaks[0]],amp[peaks[0]],color='orange')
plt.show()
```
