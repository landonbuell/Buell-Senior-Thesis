```python
"""
Landon Buell
30 December 2019
"""

import numpy as np
import matplotlib.pyplot as plt
import os
import winsound
import wave

import scipy.io.wavfile as sciowav
import scipy.signal as signal
import scipy.fftpack as fftpack

rate = 10000
```

```python
time = np.arange(0,4,1/rate)
amp = np.cos(2*np.pi*2*time)+np.sin(2*np.pi*4*time)

f1,f2 = 5,8
x1 = (4*np.cos(2*np.pi*f1*time) - 3*np.cos(2*np.pi*f2*time) )#* amp
x1 = x1/np.max(np.abs(x1))

x1 = np.sin(2*np.pi*5*time)

plt.figure(figsize=(16,8))
plt.plot(time,x1)
plt.grid()
plt.tight_layout()
plt.show()
```

```python
npts = 10000
fftdata = fftpack.fft(x=x1,n=npts)
x1_power = np.abs(fftdata)**2
x1_fspace = fftpack.fftfreq(d=1/rate,n=npts)

pts = np.where((x1_fspace>=0)&(x1_fspace<=40))
x1_fspace = x1_fspace[pts]
x1_power = x1_power[pts]

plt.figure(figsize=(16,8))
plt.plot(x1_fspace,x1_power)
plt.grid()
plt.tight_layout()
plt.show()

print("Max Freq:",np.argmax(x1_power))
```

```python

x2_power = np.roll(x1_power,-200)

plt.figure(figsize=(16,8))
plt.plot(x1_fspace,x2_power)
plt.grid()
plt.tight_layout()
plt.show()

print("Max Freq:",np.argmax(x2_power))
```
