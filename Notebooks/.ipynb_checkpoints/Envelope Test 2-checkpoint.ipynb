```python
"""
Landon Buell
30 December 2019
"""

import numpy as np
import matplotlib.pyplot as plt
import os
import scipy.signal as signal
import scipy.io.wavfile as sciowav
```

```python
t = np.arange(0,2*np.pi,0.001)
amp = np.sin(2*np.pi*t)
f = 12
waveform = amp*np.cos(2*np.pi*f*t)
```

```python
        #### AMPLITUDE ENVELOPE ####
    
analytic = signal.hilbert(waveform)
amp_env = np.abs(analytic)
```

```python
        #### PLOTTING ####

plt.figure(figsize=(20,8))
plt.title("Waveform Envelope Test",size=40,weight='bold')
plt.xlabel("Time",size=20,weight='bold')
plt.ylabel("Amplitude",size=20,weight='bold')

plt.plot(t,waveform,label='signal')
plt.plot(t,amp,label='Given Envelope')
plt.plt(t,amp_env,label='Hilbert Envelope')

plt.xticks(np.arange(0,len(t),int(1e4)))
plt.yticks([-1.00,-0.75,-0.50,-0.25,0.00,+0.25,+0.50,+0.75,+1.00])
plt.hlines(0,0,2*np.pi)

plt.grid()
plt.legend()
plt.tight_layout()

plt.show()
```
