```python
"""
Landon Buell
30 December 2019
"""

import numpy as np
import matplotlib.pyplot as plt
import os

import scipy.signal as signal
import scipy.io.wavfile as sciowav

int_dir = 'C:/Users/Landon/Documents/GitHub/Buell-Senior-Thesis/Notebooks'
wav_dir = 'C:/Users/Landon/Documents/wav_audio'
```

```python
        #### LOAD .WAV FILE ####

filename = 'Viola.arco.sulG.G4.L.wav'
os.chdir(wav_dir)
rate,waveform = sciowav.read(filename)
os.chdir(int_dir)
waveform = waveform/np.max(waveform)
indexes = np.arange(0,len(waveform),1)
time = indexes/44100
print("Points in wav file:",len(waveform))
print("Time of wav file:",len(waveform)/44100)

plt.figure(figsize=(16,4))
plt.plot(indexes,waveform,color='blue')
plt.grid()
plt.show()
```

```python

peaks,properties = signal.find_peaks(x=waveform,height=0.1,threshold=0.1,
                                        distance=4410,
                                         prominence=1.0)
                                     

plt.figure(figsize=(16,4))
plt.plot(indexes,waveform,color='blue')
plt.plot(peaks,waveform[peaks],'go')
plt.xlim(12000,125000)
plt.grid()
plt.show()
```
