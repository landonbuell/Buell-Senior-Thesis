```python
cols = ['Rise Time', 'Decay Time', 'RMS Frame Energy', 'RMS < 10% of max',
       'RMS < 25% of max', 'RMS < 50% of max', 'RMS < 75% of max', 'Avg TSF',
       'Target']
FRAME = pd.read_csv('C:/Users/Landon/Documents/GitHub/Buell-Senior-Thesis/Instrument_Classifier_v0/INST_CLF_v0/extdata/Output.csv',
                  names=cols)
FRAME.head(10)
```

```python
y = FRAME['Target']
y = y.to_numpy()
int_to_fam = {0:'Brass',1:'String',2:'High_Winds',3:'Low_Winds'}
#y = [int_to_fam[y] for y in y]
y
```

```python
def Plot_Features(DATA,X1,X2,Y,savename=''):
    """ Plot 2D feature space, color code by class """
    plt.figure(figsize=(16,12))
    plt.xlabel(str(X1),size=20,weight='bold')
    plt.ylabel(str(X2),size=20,weight='bold')
    
    plt.scatter(x=DATA[X1],y=DATA[X2],c=Y,cmap='Set1')
                
    plt.vlines(0,0,1,color='black')
    plt.hlines(0,0,1,color='black')
    #plt.legend()
    plt.grid()
    plt.tight_layout()
    plt.savefig(savename+'.png')
    plt.show()
```

```python
""" Plot RISE vs. DECAY """
Plot_Features(FRAME,'Rise Time','Decay Time',Y=y,
             savename="Rise_Decay")
```

```python
""" Plot RMS vs. RMS <10 % """
Plot_Features(FRAME,'RMS Frame Energy','RMS < 10% of max',Y=y,
             savename="RMS_10")
```

```python
""" Plot RMS vs. RMS <25 % """
Plot_Features(FRAME,'RMS Frame Energy','RMS < 25% of max',Y=y,
             savename="RMS_25")
```

```python
""" Plot RMS vs. RMS <50 % """
Plot_Features(FRAME,'RMS Frame Energy','RMS < 50% of max',Y=y,
             savename="RMS_50")
```

```python
""" Plot RMS vs. RMS <75 % """
Plot_Features(FRAME,'RMS Frame Energy','RMS < 75% of max',Y=y,
             savename="RMS_75")
```

```python
def Plot_Features2(DATA,X1,X2,Y,savename=''):
    """ Plot 2D feature space, color code by class """
    plt.figure(figsize=(16,8))
    plt.xlabel(str(X1),size=20,weight='bold')
    plt.ylabel(str(X2),size=20,weight='bold')
    
    plt.scatter(x=DATA[X1],y=DATA[X2]/DATA[X2].max(),c=Y,cmap='Set1')
                
    #plt.vlines(0,0,1,color='black')
    #plt.hlines(0,0,1,color='black')
    #plt.legend()
    plt.grid()
    plt.tight_layout()
    plt.savefig(savename+'.png')
    plt.show()
    
Plot_Features2(FRAME,'RMS Frame Energy','Avg TSF',Y=y,
             savename="RMS_TSF")
```
