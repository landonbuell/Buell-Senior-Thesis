```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.fftpack as fftpack
import time

n = (2**14)
print(n)
rate = 44100
dt = 1/rate
f1,f2 = 200,5000

t = np.arange(0,2,dt)
x1 = 2*np.sin(2*np.pi*f1*t) + 3*np.sin(2*np.pi*f2*t)
```

```python
y1_fft = fftpack.fft(x=x1,n=n)
y1 = (y1_fft)*np.conj(y1_fft)
fspace = fftpack.fftfreq(n=n,d=dt)

print(fspace)
```

```python
plt.figure(figsize=(20,8))
plt.title("Scipy Fast Fourier Transform",size=40,weight='bold')
plt.xlabel("Frequency",size=20,weight='bold')
plt.ylabel("Amplitude",size=20,weight='bold')

plt.plot(fspace,y1,color='blue')

plt.grid()
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(20,8))
plt.title("Scipy Fast Fourier Transform",size=40,weight='bold')
plt.xlabel("Frequency",size=20,weight='bold')
plt.ylabel("Amplitude",size=20,weight='bold')

pts = np.where((fspace>0)&(fspace<250))
y2 = y1[pts]
plt.plot(fspace[pts],y2,color='blue')

plt.grid()
plt.tight_layout()
plt.show()

print(len(y2))
```

```python
y3 = np.array([])
nu = np.arange(n)/n
print(len(x1))
print(len(nu))
x2 = x1[0:n]

for k in range(0,n,1):
    yk = np.sum(x2*np.exp(-2*np.pi*-1j*k*nu))
    y3 = np.append(y3,yk)
```

```python
plt.figure(figsize=(20,8))
plt.title("Scipy Fast Fourier Transform",size=40,weight='bold')
plt.xlabel("Frequency",size=20,weight='bold')
plt.ylabel("Amplitude",size=20,weight='bold')

plt.plot(fspace,y3,color='blue')

plt.grid()
plt.tight_layout()
plt.show()
```
