```python
"""
Landon Buell
Fourier Transforms
Test 0
10 January 2020
"""

import numpy as np
import matplotlib.pyplot as plt
import scipy.fftpack as fftpack

rate = 44100
dt = 1/rate
t = np.arange(0,3,dt)

""" For this experiment, I will set the sample rate to 44,100 Hz as per the Nyquist theorem
the time spaceing, dt, is the reciprocal of the sample rate
The time axis will be an even 3 seconds"""

time_pts = len(t)
print(time_pts)
```

```python
def plot_waveform (time,ydata,labs):
    """ Create matplotlib fgured of desired waveform arrays"""
    plt.figure(figsize=(20,8))
    plt.xlabel('Time',size=20,weight='bold')
    plt.ylabel('Amplitude',size=20,weight='bold')
    try:
        for I in range(len(ydata)):
            plt.plot(time,ydata[I],label=labs[I])
        plt.legend()
    except:
        plt.plot(time,ydata)
    plt.xticks(np.arange(0,3.5,0.5))
    plt.yticks(np.arange(-1,+1.25,0.25))
    plt.grid()
    plt.tight_layout()
    plt.show()
    
def plot_frequency (freq,ydata,labs):
    """ Create matplotlib fgured of desired waveform arrays"""
    plt.figure(figsize=(20,8))
    plt.xlabel('Frequency',size=20,weight='bold')
    plt.ylabel('Amplitude',size=20,weight='bold')
    try:
        for I in range(len(ydata)):
            plt.plot(freq,ydata[I],label=labs[I])
        plt.legend()
    except:
        plt.plot(freq,ydata)
    plt.grid()
    plt.tight_layout()
    plt.show()
```

```python
def FFT (data):
    """ Compute discrete Fast Fourier Transform w/ Scipy """
    fftdata = fftpack.fft(data,n=4410)
    power = np.sqrt(fftdata*np.conj(fftdata))
    return power
    
```

```python
""" Start w/ a nice, clean integer linear frequency value, f = 110 Hz, A2 """
f1 = 110
x1 = np.cos(2*np.pi*f1*t)

fft1 = fftpack.fft(x1,n=int(time_pts*0.1))
power1 = np.sqrt(fft1*np.conj(fft1))
fspace = fftpack.fftfreq(len(power1),dt)

print("Length of FFT array:",len(fft1))

pts = np.where((fspace>= +60) & (fspace <= + 160))
power1 = power1[pts]
fspace1 = fspace[pts]

""" Notice, a single spike at f1 = 110 Hz
This spike is 1 index wide - effectivley a delta function, or a very narrow gaussian"""

plot_frequency(fspace1,power1,['f1'])
```
